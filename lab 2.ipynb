```python
import cv2
import matplotlib.pyplot as plt
```

```python
print(cv2.__version__)
```

```python
image_color = cv2.imread('/Users/230108/Downloads/tulip.jpeg')
image_sub = cv2.subtract(image_color, 100)

cv2.imshow('subtracted', image_sub)
cv2.waitKey(-1)
```

```python
image = cv2.imread('/Users/230108/Downloads/tulip.jpeg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
subtracted_image = cv2.subtract(image_RGB, 100)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(subtracted_image)
plt.title('Subtracted Image (-100)')

plt.axis('off')
plt.show()
```

```python
red_only = image.copy()
red_only[:,:,0] = cv2.subtract(red_only[:,:,0],100)

plt.imshow(red_only)
plt.title('Increased Red Channel')
plt.axis('off')
plt.show()
```

```python
add_30 = cv2.add(image_RGB, 30)
add_100 = cv2.add(image_RGB, 100)

plt.figure(figsize=(9,5))
plt.subplot(1,3,1); plt.imshow(image_RGB); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(add_30); plt.title("Added 30"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(add_100); plt.title("Added 100"); plt.axis('off')
plt.show()
```

```python
div_one = cv2.multiply(image_RGB,3.0)
div_two = cv2.multiply(image_RGB, 0.5)

plt.figure(figsize=(9,5))
plt.subplot(1,3,1); plt.imshow(image_RGB); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(div_one); plt.title("Divide to 3.0 "); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(div_two); plt.title("Divide to 0.5"); plt.axis('off')
plt.show()
```
