```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/Users/230108/Downloads/greece.jpg')
print(image.shape)
```

```python
translation_matrix = np.float32([[1, 0, 100], [0, 1, 50]])
translated_image = cv2.warpAffine(image,
translation_matrix, (image.shape[1],image.shape[0]))
print(translated_image.shape)
```

```python
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title('Task 1: Negative Shift (Left & Up)')
plt.axis('off')
plt.show()
```

```python
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 90, 1.0)

rotated_image = cv2.warpAffine(image,rotation_matrix, (w, h))
```

```python
plt.imshow(cv2.cvtColor(rotated_image,
cv2.COLOR_BGR2RGB))
plt.title('Task 2: 90 Degree Rotation with 0.7 Scale')
plt.axis('off')
plt.show()
```

```python
Factor = 5
scaling_matrix = cv2.getRotationMatrix2D((0, 0), 0,Factor)

scaled_image = cv2.warpAffine(image,scaling_matrix, (w*Factor, h*Factor))
```

```python
plt.imshow(cv2.cvtColor(scaled_image,
cv2.COLOR_BGR2RGB))
plt.title('Task 3: Scaled Image 5x (Check Pixelation)')
plt.axis('off')
plt.show()
print(scaled_image.shape)
```

```python
m = 0.67
shearing_matrix = np.float32([[1, m, 0], [0, 1, 0]])

sheared_image = cv2.warpAffine(image, shearing_matrix, (w, h))
```

```python
plt.imshow(cv2.cvtColor(sheared_image, cv2.COLOR_BGR2RGB))
plt.title('Task Sheared Image (x-axis, factor ' + str(m)+')')
plt.axis('off')
plt.show()
```

```python
m_x = 0.2
m_y = 0.1

M_task4 = np.float32([[1, m_x, 0], [m_y, 1, 0]])
sheared_task4 = cv2.warpAffine(image, M_task4, (int(w*1.5), int(h*1.5)))

plt.imshow(cv2.cvtColor(sheared_task4, cv2.COLOR_BGR2RGB))
plt.title(f'Task 4: Shearing (x={m_x}, y={m_y})')
plt.axis('off')
plt.show()
```
